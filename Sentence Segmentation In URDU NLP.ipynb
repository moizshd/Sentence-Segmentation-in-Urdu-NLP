#### Accuracy

##### For Single Paragraph

```
def acc(test_text,label_text):
  correct=0
  for i in label_text:
    for j in test_text:
      lb=i.split(' ')
      ts=j.split(' ')
      if(len(lb)>2 and len(ts)>2):
        if(lb[-1]==ts[-1] and lb[-2]==ts[-2] and lb[-3]==ts[-3]):
          correct=correct+1
      else:
        return('Error')
  accuracy=(float(correct)/float(len(label_text)))*100
  # print('Accuracy\t=\t'+str(accuracy)+'%')
  return accuracy

```

##### For Multiple Paragraph

```
def Accuracy(train,test):

  total=0
  e=0
  for i in range(len(test)):
    train_text = sentence_tokenizer(train[i][0])
    label_text = test[i][0].split('۔')
    tmp=acc(train_text,label_text)
    if(tmp!='Error'):
          total=total+tmp
    else:
      e=e+1

  accuracy=(float(total)/float(len(test)-e))
  print('Accuracy\t=\t'+str(accuracy)+'%')

```

```
def conv_tokens(paragraph):
  nlp = spacy.blank('ur')
  doc = nlp(paragraph)
  tokens=[]
  for word in doc:
      tokens.append(word)
  return tokens
```

```
URDU_EOS_SET = ['کیجیے', 'کیجئے', 'گئیں', 'تھیں', 'ہوں', 'خریدا', 'گے', 'ہونگے', 'گا', 'چاہیے', 'ہوئیں', 'گی',
                      'تھا', 'تھی', 'تھے', 'ہیں', 'ہے','ہوگا', 'کہا', 'جاسکے', 'رہی','کیں','گیا','دیا','رکھتے','گئی','لگے','لگی','لگا','جائے'
                       ] 
URDU_CONJUNCTIONS = ['جنہیں', 'جس', 'جن', 'جو', 'اور','اگر','اگرچہ', 'لیکن', 'مگر', 'پر', 'یا', 'تاہم', 'کہ', 'کر',
                      'تو', 'گے', 'گی','حالانکہ', 'لہذا', 'باوجود اس کے ', 'کیونکہ', 'جاتا','تاکہ', '،', 'بقول','جسے','جیسے','ہو','ورنہ','اتنا','اتنی','بلکہ',
               'اسے',  ':', '؟','دوسری'  ,'چونکہ'
                      ]
```

```
def convert_spacy_tokens_to_str_tokens(testwords):
  mytext=[]
  for i in testwords:
    mytext.append(i.text)
  return mytext
```

```
def index_func(tokenized_string):
  index1=[]
  temp=[]
  final_index=[]
  for i in tokenized_string:
    for j in URDU_EOS_SET:
      if i.text==j:
        ind=(tokenized_string.index(i))
        index1.append(ind)
  for i in index1:
    for k in URDU_CONJUNCTIONS:
      if i+1<len(tokenized_string):
        if tokenized_string[i+1].text==k:
          temp.append(i)
  for i in index1:
    for j in URDU_EOS_SET:
      if i+1<len(tokenized_string):
        if tokenized_string[i+1].text==j:
          temp.append(i)
  for element in index1:
      if element not in temp:
          final_index.append(element)
  return final_index
  
```

```
def fillperiods(tokenized_words,indexes):
  itr=1
  for i in indexes:
    tokenized_words.insert(i+itr,'-')
    itr=itr+1
  return tokenized_words
def jointokens(tokenized_words):
  text=' '.join(tokenized_words)
  print(text)
  return text
```
